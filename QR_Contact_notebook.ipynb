```python
import segno
import vobject
from IPython.display import SVG, display, Image
import IPython.display as dis
from urllib.request import urlopen
```

```python
url = "https://lirp.cdn-website.com/aa0ef369/dms3rep/multi/opt/google-cloud-icon-400w.png"
bq_file = urlopen(url)
```

```python
first_name = f"Dennis"
last_name = f"YUE"
email = f"dennisyue@google.com"
title = f"Head of Sales"
org = f"Google Cloud Hong Kong"
mobile = f"+85255058310"
work = f"+85239235586"
website = f"https://cloud.google.com/"
```

```python
content = F"BEGIN:VCARD \n\
VERSION:3.0 \n\
FN;CHARSET=UTF-8:{first_name} {last_name} \n\
N;CHARSET=UTF-8:{last_name};{first_name};;; \n\
EMAIL;CHARSET=UTF-8;type=WORK,INTERNET:{email} \n\
TEL;TYPE=CELL:{mobile} \n\
TEL;TYPE=WORK,VOICE:{work} \n\
TITLE;CHARSET=UTF-8:{title} \n\
ORG;CHARSET=UTF-8:{org} \n\
URL;type=WORK;CHARSET=UTF-8:{website} \n\
REV: \n\
END:VCARD \n"
```

```python
qr = segno.make(content=content)
```

```python
qr.to_artistic(background=bq_file, target='vcard.png', scale=4)
```

```python
dis.Image(filename='vcard.png')
```
