```python
import segno
from segno import helpers
from IPython.display import SVG, display, Image
import IPython.display as dis
from urllib.request import urlopen
```

```python
qr = segno.make_mecard(name='Thomas, Qian', email='thomasqian@google.com', phone='+66189831')
```

```python
qr.designator
```

```python
qr.save('my-mecard.svg', scale=4)
```

```python
display(SVG(filename="/my-mecard.svg"))
```

```python
qr = helpers.make_vcard(name='Qian;Thomas', 
                        displayname='Tom $', 
                        email='thomasqian@google.com', 
                        phone='+85266189831',
                        title='AI Specialist', 
                        org="Google Cloud Platform",
                        url=("https://cloud.google.com/products/ai","8.8.8.8")
                       )
```

```python
qr.save('my-vcard.svg', scale=4)
```

```python
display(SVG(filename="/my-vcard.svg"))
```

```python
url = "https://lirp.cdn-website.com/aa0ef369/dms3rep/multi/opt/google-cloud-icon-400w.png"
bq_file = urlopen(url)
```

```python
qr.to_artistic(background=bq_file, target='vcard.gif', scale=2)
```

```python
dis.Image(filename='vcard.gif')
```
